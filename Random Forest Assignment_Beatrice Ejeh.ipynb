# Iris Dataset

```python
#Import the required libaries
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#Use the dir function to return all properties and methods of Iris
dir(iris)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Create a DataFrame for the Iris Dataset
df_iris = pd.DataFrame(iris.data)
df_iris.head()
```

```python
# Define the target for the Iris Dataset
df_iris["target"] = iris.target
```

```python
df_iris[0:12]
```

```python
#Defining both X(feature) and y(target) for the Iris dataset
X = df_iris.drop("target", axis = "columns")
y = df_iris.target
```

```python
#Train and test the Iris dataset using the train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
#Instantiate and fit the RandomForest Model
from sklearn.ensemble import RandomForestClassifier
model_iris = RandomForestClassifier (n_estimators = 20)
model_iris.fit(X_train, y_train)
```

```python
y_test
```

```python
X_test
```

```python
#Check the model score
model_iris.score(X_test, y_test)
```

```python
# Model Prediction
y_pred_value = model_iris.predict(X_test)
y_pred_value
```

```python
#Checking the Confusioin matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred_value)
confusion_matrix
```

```python
#Plotting the confusion matrix
import seaborn as sns
plt.figure(figsize = (8,5))
sns.heatmap(confusion_matrix, annot = True)
plt.xlabel("Predicted Value")
plt.ylabel("Truth")
```

# Diabetes Dataset

```python
#Import the Diabetes Dataset from Sklearn
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
#Use the dir function to check the properties and methods of the dataset
dir(diabetes)
```

```python
#Transform the diabetes dataset into a Pandas DataFrame
df_diabetes = pd.DataFrame(diabetes.data)
df_diabetes.head()
```

```python
#Create the target
df_diabetes["target"] = diabetes.target
```

```python
df_diabetes[0:12]
```

```python
#Define the X(feature) and y(target) of the diabetes dataset
X = df_diabetes.drop("target", axis = "columns")
y = df_diabetes.target
```

```python
#Train, test and split the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.35)
```

```python
#Instantiate and fit the model using RandomForest
from sklearn.ensemble import RandomForestClassifier
model_diabetes = RandomForestClassifier (n_estimators = 20)
model_diabetes.fit(X_train, y_train)
```

```python
y_test
```

```python
X_test
```

```python
#Check the accuracy of the model
model_diabetes.score(X_test, y_test)
```

```python
#Model Prediction
y_pred_val = model_diabetes.predict(X_test)
y_pred_val
```

```python
#Confusion matrix for the diabetes dataset
from sklearn.metrics import confusion_matrix
confusion_mat = confusion_matrix(y_test, y_pred_val)
confusion_mat
```

```python
#Plot the confusion matrix
import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(confusion_mat, annot = True)
plt.xlabel("Predicted Value")
plt.ylabel("Truth")
```
